```python
import pandas as pd

# Load your dataset (update path if needed)
df = pd.read_csv(r"C:\\Users\\Vishnu\\Desktop\\data_cleaning_prod\\Data_files\SATARA_1991.csv")

# Quick check
print(" Dataset loaded successfully!")
print(f"Shape: {df.shape}")
df.head()
```

```python
# Calculate Data Quality Score
total_rows, total_cols = df.shape
total_cells = total_rows * total_cols
missing_cells = df.isnull().sum().sum()
data_quality_score = round((1 - (missing_cells / total_cells)) * 100, 2)

print(f"Data Quality Score: {data_quality_score}%")

#  Basic interpretation
if data_quality_score > 95:
    interpretation = "Excellent — almost no missing data."
elif data_quality_score > 85:
    interpretation = "Good — small number of missing values, manageable."
elif data_quality_score > 70:
    interpretation = "Moderate — needs cleaning or imputation."
else:
    interpretation = "Poor — significant missing data, consider preprocessing."
    
print("Interpretation:", interpretation)
```
